### TASK 2 (2 балла)

Реализуйте класс EM-алгоритма для GMM, не пользуясь готовым решением из sklearn

* В классе обязательно должны быть методы вызова E-шага и M-шага
* На каждом шаге мы должны логировать обновленные параметры, пересчитанную апостериорную вероятность принадлежности к кластеру и значение правдоподобия

* Постайтесь написать сами, все-таки модель смеси самая простая для понимания и реализации

```python
from scipy.stats import multivariate_normal

class MY_GMM:
    def __init__(self, K, max_iter=100, tol=1e-6):
        self.K        = K
        self.max_iter = max_iter
        self.tol      = tol

    def _init_params(self, X):
        _, d = X.shape
        self.pi_    = np.ones(self.K) / self.K 
        self.means_ = X[:self.K]   
        self.covs_  = np.array([np.eye(d) for _ in range(self.K)])

    def _get_all_probas(self, X):
        prob = np.array([
            multivariate_normal(mean=self.means_[k], cov=self.covs_[k], allow_singular=True).pdf(X)
            for k in range(self.K)
        ]).T
        return prob

    def _compute_log_likelihood(self, X):
        all_probs = self._get_all_probas(X)

        log_likelihood = np.sum(np.log(all_probs @ self.pi_))

        return log_likelihood

    def e_step(self, X):
        all_probs = self._get_all_probas(X)

        g_numerator = all_probs * self.pi_
        normalizetion_contants = g_numerator.sum(axis=1, keepdims=True)

        g_posterior = g_numerator / normalizetion_contants

        return g_posterior

    def m_step(self, X, g):
        n, d = X.shape

        unnormalized_pi = g.sum(axis=0)
        
        self.pi_ = unnormalized_pi / n
        
        self.means_ = (g.T @ X) / unnormalized_pi[:, None]
        
        self.covariances_ = np.zeros((self.K , d, d))
        
        for k in range(self.K):
            diff = X - self.means_[k]
            self.covs_[k] = (g[:, k, None] * diff).T @ diff / unnormalized_pi[k]
        
        return self.pi_, self.means_, self.covs_
    
    def fit(self, X):      

        self._init_params(X)
        
        history = []
        prev_log_likelihood = - np.inf
        
        for _ in range(self.max_iter):
            
            # E-step
            g = self.e_step(X)
            
            log_likelihood = self._compute_log_likelihood(X)
            history.append(log_likelihood)
            
            if abs(log_likelihood - prev_log_likelihood) < self.tol:
                break
        
            # M-srep
            self.m_step(X, g)
            
            prev_log_likelihood = log_likelihood

        self.history = history
        
        return history
    
    def predict_proba(self, X):
        g = self.e_step(X)
        return g
    
    def predict(self, X):
        g = self.predict_proba(X)
        return np.argmax(g, axis=1)
```

### TASK 3 (2 балла)

Докажите, что на каждом шаге EM-алгоритма значение правдоподобия не уменьшается

* Здесь, скорее всего, придется воспользоваться понятиями KL-дивергенции и ELBO
* Обязательно прокомментируйте, откуда взялись эти термины
* Проверяющий должнен понять, что Вы разобрались в новых концепциях (не просто скатали)

Вспомним вывод EM алгоритма:

$
\log p(x | \theta) = \int \log p(x | \theta) q(z) dz = 
\int \log \left [p(x | \theta) \frac{q(z)}{q(z)}\right] q(z) dz = \int \log \left [\frac{p(x, z| \theta)}{p(z |x, \theta)} \frac{q(z)}{q(z)}\right] q(z) = \int \log \left [\frac{p(x, z| \theta)}{q(z)} \frac{q(z)}{p(z | x, \theta)}\right] q(z) = \underbrace{\int \log \frac{p(x, z| \theta)}{q(z)}q(z)dz}_{\mathcal{L(q, \theta)}} +  \underbrace{\int \log \frac{q(z)}{p(z | x, \theta)} q(z)dz}_{D_{KL}(q(z), p(z | x, \theta))}
$

Тем самым получили:
$$\log p(x | \theta) = \mathcal{L}(q, \theta) + D_{KL}(q(z), p(z | x, \theta)),$$
где $\mathcal{L}(q, \theta)$ - ELBO, $D_{KL}(q(z), p(z | x, \theta))$ - дивергенция Кульбака-Лейблера.

- Почему теперь правдоподобие на каждом шаге не уменьшается?

Пусть на шаге $t$  имеем параметры  $\theta_t$.

- **E-шаг (из определения):**  
$
q_{t+1}(z) = p(z | x, \theta_t).
$

При этом $ D_{KL}(q_{t+1} \,\|\, p(z|x,\theta_t)) = 0 $ (по свойству дивергенции Кульбака-Лейблера).

Тогда
$
\log p(x|\theta_t) = \mathcal{L}(q_{t+1}, \theta_t).
$

- **M-шаг (из определения):**  
Фиксируем $ q_{t+1}(z) $, максимизируем $ \mathcal{L}(q_{t+1}, \theta)$ по $\theta$:
  
$
\theta_{t+1} = \arg\max_{\theta} \mathcal{L}(q_{t+1}, \theta) \Rightarrow \mathcal{L}(q_{t+1}, \theta_{t+1}) \ge \mathcal{L}(q_{t+1}, \theta_t)
$  

Итого:

$$
\boxed{\log p(x| \theta_t) = \mathcal{L}(q_{t+1}, \theta_t) \leq \mathcal{L}(q_{t+1}, \theta_{t+1}) \leq \log p(x|\theta_{t+1})}
$$

Ч.т.д.

Последнее неравенство верно, так как 

$D_{KL}(q_{t+1} \,\|\, p(z|x,\theta_t)) \geq 0$ (по свойству дивергенции Кульбака-Лейблера) $\Rightarrow \log p(x | \theta) \geq \mathcal{L}(q, \theta)$

Для выбора оптимального количества кластеров будем отслеживать две метрики - инерции и силуэта (самая информативная). Силуэт скор долго считается для больших наборов, так что будем считать и усреднять по подвыборкам из всего датасета.

```python
from tqdm import tqdm
```

```python
from sklearn.metrics import silhouette_score

n_runs = 10
sample_size = 5000
k_values = range(2, 6)

sil_results = np.zeros((len(k_values), n_runs))
ll_results = np.zeros((len(k_values), n_runs))
inertia_results = np.zeros((len(k_values), n_runs))


for i, k in enumerate(tqdm(k_values)):
    model = MY_GMM(K=k, max_iter=1_000, tol=1e-5)
    model.fit(X_pca)
    
    for j in range(n_runs):
        indices = np.random.choice(len(X_pca), size=sample_size, replace=False)
        X_sample = X_pca[indices]
        labels = model.predict(X_sample)
        
        sil_results[i, j] = silhouette_score(X_sample, labels)
        ll_results[i, j] = model._compute_log_likelihood(X_sample) / sample_size
        
        centers = model.means_
        dists = np.linalg.norm(X_sample - centers[labels], axis=1) ** 2
        inertia_results[i, j] = np.sum(dists)

sil_mean = sil_results.mean(axis=1)
sil_std = sil_results.std(axis=1)
ll_mean = ll_results.mean(axis=1)
ll_std = ll_results.std(axis=1)
inertia_mean = inertia_results.mean(axis=1)
inertia_std = inertia_results.std(axis=1)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))

ax1.plot(k_values, sil_mean)
ax1.fill_between(k_values, sil_mean - sil_std, sil_mean + sil_std, alpha=0.2)
ax1.set_xlabel('K')
ax1.set_ylabel('Silhouette')
ax1.grid(True)

ax2.plot(k_values, ll_mean)
ax2.fill_between(k_values, ll_mean - ll_std, ll_mean + ll_std, alpha=0.2)
ax2.set_xlabel('K')
ax2.set_ylabel('Log-Likelihood')
ax2.grid(True)

ax3.plot(k_values, inertia_mean)
ax3.fill_between(k_values, inertia_mean - inertia_std, inertia_mean + inertia_std, alpha=0.2)
ax3.set_xlabel('K')
ax3.set_ylabel('Inertia')
ax3.grid(True)

plt.tight_layout()
plt.show()
```

Правдоподобие растет при увеличении количества кластеров, что логично (эта величина отслеживается для лучшего контроля).

Самая информативная метрика - силуэт скор - падает с ростом количества кластеров на значениях K > 4, но максимум имеет при K = 4. По методу локтя, опираясь на третий график, также видно, что нужно выбрать K = 4.

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(X_pca)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, s=1, alpha=0.1)
ax1.set_title('GMM')
ax1.set_xlabel('PCA 0')
ax1.set_ylabel('PCA 1')

ax2.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans_labels, s=1, alpha=0.1)
ax2.set_title('K-Means')
ax2.set_xlabel('PCA 0')

plt.tight_layout()
plt.show()
```

Визуальный анализ подтверждают, что результаты методов кардинально отличаются. K-Means нарезал пространство на сектора прямыми линиями, игнорируя реальное распределение плотности, тогда как GMM выделил более естественные эллиптические области. Это доказывает, что структура данных сложнее простых сфер, и GMM справляется с ней лучше.

Теоретическая связь методов заключается в том, что K-Means является частным случаем GMM. Он получается, если наложить на GMM два ограничения: зафиксировать форму всех кластеров как идеальные сферы (единичная ковариация) и запретить вероятностную принадлежность, требуя жесткой привязки каждой точки только к одному центру (вероятность 0 или 1).

Придумайте несколько здравых гипотез о датасете и проверьте их с помощью параметрических и непараметрических критериев

Обязательно должно быть по одной гипотезе каждого вида:

Сравните результаты, каким тестам в данной задаче Вы доверяете больше?

```python
from scipy import stats

df_clean = df.dropna(subset=num_cols + cat_cols + ['Churn']).copy()
df_test_clean = df_test.dropna(subset=num_cols + cat_cols + ['Churn']).copy()

churn_0 = df_clean[df_clean['Churn'] == 0]
churn_1 = df_clean[df_clean['Churn'] == 1]
```

### Гипотеза о сдвиге: проверяем, отличается ли средний возраст ушедших клиентов от лояльных.

H0: Средние в двух группах равны.

H1: Средние различаются.

```python
t_stat, p_val_t = stats.ttest_ind(churn_0['Age'], churn_1['Age'], equal_var=False)
u_stat, p_val_u = stats.mannwhitneyu(churn_0['Age'], churn_1['Age'])

print(f"T-test p-value: {p_val_t:.4e}")
print(f"Mann-Whitney p-value: {p_val_u:.4e}")
print(f"Средний возраст: Оставшихся - {churn_0['Age'].mean():.2f}, Ушедших - {churn_1['Age'].mean():.2f}")
```

#### Вывод: 
В рассмотренных группах существует сдвиг - средний возраст ушедших (42) больше среднего возраста оставшихся (36)

### Гипотеза о масштабе: проверяем, отличается ли разброс (дисперсия) трат у лояльных и ушедших клиентов.

H0: Дисперсии групп одинаковы.

H1: Дисперсии различаются.

```python
stat_b, p_val_b = stats.bartlett(churn_0['Total Spend'], churn_1['Total Spend'])
stat_l, p_val_l = stats.levene(churn_0['Total Spend'], churn_1['Total Spend'])

print(f"Bartlett p-value: {p_val_b:.4e}")
print(f"Levene p-value: {p_val_l:.4e}")
print(f"Std Dev: Оставшиеся - {churn_0['Total Spend'].std():.2f}, Ушедшие - {churn_1['Total Spend'].std():.2f}")
```

#### Вывод: 
В рассмотренных группах наблюдается различие в масштабе трат - дисперсия трат ушедших (260) больше дисперсии трат оставшихся (144)

### Гипотеза о связи: ищем взаимосвязь между частотой использования (Usage Frequency) и задержкой оплаты (Payment Delay).

H0: Корреляция равна 0 (связи нет).

H1: Корреляция не равна 0.

```python
corr_p, p_val_p = stats.pearsonr(df_clean['Usage Frequency'], df_clean['Payment Delay'])
corr_s, p_val_s = stats.spearmanr(df_clean['Usage Frequency'], df_clean['Payment Delay'])

print(f"Pearson correlation: {corr_p:.4f} (p={p_val_p:.4e})")
print(f"Spearman correlation: {corr_s:.4f} (p={p_val_s:.4e})")
```

#### Вывод: 

В рассмотренных группах не наблюдается линейной зависимости между частотой использования и задержками в оплате
